# 전화번호 프로그램 만들기

* 홍길동 010-123-4567
* 기능1 : 연락처 추가
* 기능2 : 연락처 전체 보기
* 기능3 : 검색, 이름을 입력받아서 전화번호 조회
* 기능4 : 수정, 이름을 입력받아서 전화번호 입력수정
* 기능5 : 삭제, 이름 입력받아서 삭제
* 기능6 : 프로그램 종료

```python
list ={}


while True:
    print("---------------           연락처 목록       -------------------------")
    print("Options :1.연락처추가 2.연락처목록 3.검색 4.연락처수정 5.삭제  0.종료")
    
    opt=input("옵션 번호 : ")
    
    if opt == "1":
        print("연락처를 추가합니다")
        new_name=input("이름 추가: ")
        if new_name not in list:
            list.setdefault(new_name, new_num) 
            new_num=input("전화번호 추가 : ") #setdefault ? 함수
        else :
            print("등록된 이름입니다")
         
    
    elif opt == "2":
        print("연락처 목록을 조회합니다")
        for name, tel in list.items():   #.items() 딕의 두값 모두
            print(name, tel)
        
    elif opt == "3":
        print("전화번호를 조회합니다")
        name=input("이름 입력 : ")
        if name not in list:
            print("등록되지 않은 이름입니다") 
        else: 
            print(list[name])
        
    elif opt == "4":
        print("전화번호를 수정합니다")
        name1=input("이름 입력 : ")
        update=input("수정할 전화번호 : ")
        list[name1]=update      ##딕 키값으로 인덱스 해서 수정
                    
    elif opt == "5" :
        print("연락처를 삭제합니다")
        name2=input("이름 입력 : ")
        if name2 in list:
            che=input("정말 삭제하시겠습니까 v/n ")
            if che == "y":
                    del list[name2]
        else : print("등록되지 않은 이름입니다")
        
    elif opt == "0":
         print("종료됩니다")
         break
        
    else :
        print("없는 옵션입니다")
      
```
